## 5.2 Langkah-langkah Praktikum


```python
import pandas as pd
import numpy as np
```

```python
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)
df.head()
```

## 5.3 Pertanyaan Latihan


### Q1


```python
print(df.shape)
print(df.columns.tolist())
```

### Q2


```python
missing_jumlah = df.isnull().sum()
missing_proporsi = (df.isnull().mean() * 100).round(2)
print(pd.DataFrame({
    "missing_jumlah": missing_jumlah,
    "missing_proporsi_persen": missing_proporsi
}))
```

### Q3


```python
persen_selamat = df["Survived"].mean() * 100
print(f"{persen_selamat:.2f}%")
```

### Q4


```python
wanita_kelas1 = df[(df["Sex"] == "female") & (df["Pclass"] == 1)]
print(len(wanita_kelas1))
print(wanita_kelas1["Age"].mean())
```

### Q5


```python
keselamatan_per_kelas = df.groupby("Pclass")["Survived"].mean()
print(keselamatan_per_kelas)
```
